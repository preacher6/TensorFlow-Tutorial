```python
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

```python
func = lambda th: np.sin(1/2*th[0]**2 - 1/4*th[1]**2 +3) * np.cos(2*th[0] + 1 - np.e**th[1])
```

```python
resol = 100
_X = np.linspace(-2, 2, resol)
_Y = np.linspace(-2, 2, resol)
```

```python
_Z = np.zeros((len(_X), len(_Y)))

for ix, valorx in enumerate(_X):
    for iy, valory in enumerate(_Y):
        _Z[iy, ix] = func([valorx, valory])
```

```python
theta = np.random.rand(2)*4 - 2
plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()
plt.plot(theta[0], theta[1], "o", c='white')
plt.show()
theta_c = theta.copy()
h = 0.001
lr = 0.001
for iter in range(1000):
    theta_c+=h
    grad = (func(theta_c) - func(theta)) / h
    theta = theta - lr * grad
    if iter%100==0:
        plt.plot(theta[0], theta[1], ".", c='red')

plt.plot(theta[0], theta[1], "o", c='green')
```

```python
plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()
plt.plot(theta[0], theta[1], "o", c='red')
plt.show()
```
